```python
import pandas as pd
import os
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
from matplotlib import pyplot as plt
from pandas.plotting import lag_plot
from pandas.plotting import autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from statsmodels.tsa.ar_model import AutoReg
```

```python
spreadPath = '/content/drive/MyDrive/Colab Notebooks/Independent project_2023 Summer/spreadData'
spreadFiles = os.listdir(spreadPath)
spreadFiles.remove('.ipynb_checkpoints')
```

```python
spreadFiles[1].split('_')[0]
```

```python
for file in spreadFiles:
  series = pd.read_csv(spreadPath + '/' + file, header=0, index_col=0)
  name = file.split('_')[0]
  series.plot()
  plt.title(name)
  plt.show()
  lag = lag_plot(series)
  plt.title(name + "_lag")
  plt.show()
  pacf = plot_pacf(series, lags=50)
  plt.title(name + "_PACF")
  plt.show()
  model = AutoReg(series, lags=1).fit()
  resi = model.resid
  plt.hist(resi, bins=300)
  plt.title(name + '_resi')
  plt.show()
```
