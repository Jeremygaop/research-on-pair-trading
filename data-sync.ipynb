# Find Common day

```python
stock3443_dates = np.unique(stock3443.date)
stoD=pd.to_datetime(stock3443_dates, format="%Y-%m-%d")
qqqq=stoD.year*10000+stoD.month*100+stoD.day
```

```python
qqqq
```

```python
len(qqqq)
```

```python
JBF_dates = np.unique(JBF.date)
indD=pd.to_datetime(JBF_dates, format="%Y-%m-%d")
pppp=indD.year*10000+indD.month*100+indD.day
```

```python
len(pppp)
```

```python
commonDays=pd.to_datetime(pppp.intersection(qqqq),format="%Y%m%d")
```

```python
commonDays
```

```python
##With common days generated for both stock and futures, we synchronize days first
d_futures=pd.to_datetime(JBF.date,format="%Y-%m-%d")
JBF.date=d_futures
JBF=JBF[JBF.date.isin(commonDays)]

d_stock=pd.to_datetime(stock3443.date,format="%Y-%m-%d")
stock3443.date=d_stock
stock3443=stock3443[stock3443.date.isin(commonDays)]
```

```python
stock3443
```

```python
JBF
```

```python
## In order to synchronize trade time, too, we create indexes for both dataframes that have the same format
stockData_DateTime = pd.to_datetime(stock3443.date.astype(str) + ' ' + stock3443.time.astype(str), format="%Y-%m-%d %H%M%S%f")
futuresData_DateTime = pd.to_datetime(JBF.date.astype(str) + ' ' + JBF.time.astype(str), format="%Y-%m-%d %H%M%S%f")
```

```python
futuresData_DateTime
```

```python
stockData_DateTime
```

```python
##N ow, we create indexes of the same format for both the stock data and the index futures data
stock3443.index = stockData_DateTime
stock3443 = stock3443[~stock3443.index.duplicated(keep='last')]  # 

JBF.index = futuresData_DateTime
JBF = JBF[~JBF.index.duplicated(keep='last')]
```

```python
## Now, we synchronize the timestamps of the two time series; here, we downsample the time series with more time stamps
### First, we union the indexes
new_index1=stock3443.index.union(JBF.index)
new_index=np.unique(new_index1)
```

```python
len(new_index),len(stock3443.index),len(JBF.index),len(new_index)-len(JBF.index)
```

```python
###Next, we insert stock time stamps to index futures data
resampledStockData = stock3443.reindex(new_index)
```

```python
resampledStockData
```

```python
###Then, we forward fill nan's in the index futures data so that date and time will not be nan's
resampledStockData.fillna(method='ffill',inplace=True)
```

```python
resampledStockData
```

```python
###down sample index data to synchronize with stock data
stockData_downsampled=resampledStockData.loc[JBF.index]
```

```python
gamma = stockData_downsampled['lastPx'][63] / JBF['midQ'][63]
```

```python
spread = np.log(stockData_downsampled['lastPx'][63:]) - gamma * np.log(JBF['midQ'][63:])
```

```python
#plot the midQ
plt.figure(figsize=(20,10))
plt.plot(spread)
```

```python
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
pacf = plot_pacf(spread, lags=10)
plt.title("PACF")
pacf.show()
```
